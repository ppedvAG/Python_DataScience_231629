## Matplotlib Grundlagen

Es gibt in Matplotlib 2 Modi um Diagramme zu zeichnen

- Globaler Modus
    - Hier werden die Eigenschaften programmiert, und danach kann show() verwendet um den Plot zu zeichnen. Im Anschluss werden die Eigenschaften wieder verworfen für den nächsten Plot
- Objektorientierter Modus
    - Hier werden der Plot und die Achsen separat gespeichert
    - Danach können der Plot und die Achsen individuell konfiguriert werden

Globaler Modus:

```python
x = np.arange(100)
plt.figure(figsize=(12, 6))
plt.title("Linien")
plt.plot(x)
plt.plot(x*2)
plt.plot(x*3)
plt.show()
```

Objektorientierter Modus:

```python
graph, achsen = plt.subplots(figsize=(12, 6))
```

```python
achsen.plot(x, 
           color="red",
           linewidth=3,
           marker='o',
           markersize=8,
           label='X')

achsen.plot(x * -1,
           label='-X')

achsen.set_xlabel("X")
achsen.set_ylabel("Y")
achsen.set_title("Ein Plot")

achsen.legend()

graph
```

```python
graph, ((a1_1, a1_2), (a2_1, a2_2)) = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))
```

```python
a1_1.plot(np.random.randint(0, 100, size=(100)), c="red")
a1_2.plot(np.random.randint(0, 100, size=(100)), c="blue")
a2_1.plot(np.random.randint(0, 100, size=(100)), c="yellow")
a2_2.plot(np.random.randint(0, 100, size=(100)), c="green")
```

## Bar Chart (Balkendiagramm)

Bar benötigt 2 Parameter: Anzahl x Elemente, Die x Elemente selbst

```python
x = np.array([np.random.randint(0, 10) for i in range(0, 10)])
y = np.array([np.random.randint(0, 10) for i in range(0, 10)])
```

```python
plt.bar(np.arange(len(x)), x,
       color="#00FF00",
       label="X Werte",
       width=0.4)
plt.bar(np.arange(len(y)) + 0.4, y,
       color="#0000FF",
       label="Y Werte",
       width=0.4)
plt.grid(color="000000", alpha=0.2, linestyle="--", axis="y", linewidth=2)
plt.xlabel("Schritte")
plt.ylabel("Werte")
plt.legend()
```

## Histogramm

Zeigt Datenpunkte kumuliert (Anzahl) in einem Array an

```python
h = np.array([np.random.randint(0, 1000) for i in range(0, 1000)])
```

```python
plt.figure(figsize=(20, 6))
plt.hist(h,
        bins=100,
        edgecolor="#BBBBBB",
        color="#00AA00")  # bins: Anzahl Gruppen

None  # Technischen Output verbergen
```

```python
plt.pie(x,
       radius=2,
       wedgeprops={"linewidth": 1, "edgecolor": "white"},  # Wedge: Jeder Teil des Diagramms ist ein Wedge
       frame=True) 

None
```
