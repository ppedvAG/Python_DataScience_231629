## Shortcuts

(Im Zellenbearbeitungsmodus)

Pfeil Hoch, Runter: Zwischen Zellen wechseln

Enter: Zelle betreten

Escape: Zelle verlassen (ohne sie auszuführen)

Shift + Enter: Zelle ausführen (funktioniert auch wenn nicht im Code)

A, B: Zellen einfügen (darüber und darunter)

D + D: Zelle löschen

M, Y: Zwischen Code und Markdown umschalten

X, C, V: Cut, Copy, Paste (ohne Strg)

Tab: Codevervollständigung
