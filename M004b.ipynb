## Andere Datenquellen

Es ist m√∂glich, auch andere Datenquellen als CSV zu benutzen (z.B. SQL oder Json von einer API)

```python
import pandas as pd
import requests as api
```

```python
json = api.get("https://open.er-api.com/v6/latest/USD")
```

```python
json.json()
```

```python
json.json()["rates"]
```

```python
readJson = pd.DataFrame.from_dict(json.json()["rates"], orient="index")
readJson.rename(columns={0: "Rates"})
```

```python
pd.read_json("https://open.er-api.com/v6/latest/USD")
```

## Daten aus der Datenbank

```python
import sqlite3 as sql
connection = sql.connect("Test.db")

cursor = connection.cursor()

cursor.execute("create table test(id integer primary key, name varchar(30))")
cursor.execute("insert into test (id, name) values(1, 'Max1')")
cursor.execute("insert into test (id, name) values(2, 'Max2')")
cursor.execute("insert into test (id, name) values(3, 'Max3')")

data = pd.read_sql("SELECT * FROM test", con=connection)
```
